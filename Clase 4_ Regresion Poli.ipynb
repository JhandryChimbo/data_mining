```python
# Importar librerías necesarias
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
```

```python
# Configuración de estilo
sns.set(style="whitegrid")
%matplotlib inline
```

```python
# 1. Generar un conjuntod de datos sintéticos no lineales
X = np.random.rand(100,1)*10
y = 3-1.5*X+0.1*X**2-0.01*X**3 + np.random.rand(100,1)*2
```

```python
# 2. Visualizar los datos
plt.scatter(X, y, label='Datos')
plt.title('Datos Sintéticos No Lineales -  Regresion Cubica')
```

```python
# 3. Transforma las caracteristicas de terminos polinomiales (Grado 3)
poly_feature = PolynomialFeatures(degree=3, include_bias=False)
X_poly = poly_feature.fit_transform(X)
```

```python
# 4. Ajustar el modelo de regresión 
model = LinearRegression()
model.fit(X_poly, y)
```

```python
# 5. Coeficiente de la ecuacion
print('Coeficientes:', model.coef_)
print('Intercepto:', model.intercept_)
print(f"y = {model.intercept_[0]:.2f} + {model.coef_[0][0]:.2f}*X + {model.coef_[0][1]:.2f}*X^2 + {model.coef_[0][2]:.2f}*X^3")
```

```python
# 6. Predicciones
y_pred = model.predict(X_poly)
```

```python
# EVALUACION DE ERROR 
mse=mean_squared_error(y,y_pred) # calcular el error cuadratico medio
print(f"Error cuadratico medio: {mse}")
```

```python
#GRAFICA CON LA VISUALIZACION DEL AJUSTE 
plt.scatter(X,y,color='blue',label='Datos reales')
plt.scatter(X,y_pred,color='red',label='Datos predichos')
plt.title('Datos Sinteticos No Lineales')
plt.xlabel('x')
plt.ylabel('y')
plt.show
```
