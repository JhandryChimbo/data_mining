```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# 1. Generar un conjuntod de datos sint√©ticos no lineales
X = np.random.rand(200,1)*2
y = 3-1.5*X+0.1*X**2-0.01*X**3 + 1.17*X**4 + 2*X**5 + np.random.rand(200,1)*12
```

```python
# 2. Visualizar los datos
plt.scatter(X, y,)
plt.title('Datos Sint√©ticos No Lineales')
plt.xlabel('X')
plt.ylabel('y')
plt.grid(True)
plt.show()
```

```python
# 3. Transforma las caracteristicas de terminos polinomiales (Grado 5)
poly = PolynomialFeatures(degree=5)
X_poly = poly.fit_transform(X)
```

```python
# 4. Ajustar el modelo de regresi√≥n 
model = LinearRegression()
model.fit(X_poly, y)
```

```python
# 5. Coeficientes de la ecuacion
print('Coeficientes:', model.coef_)
print('Intercepto:', model.intercept_)
coefficients = model.coef_.flatten() 
print(f"y = {model.intercept_[0]:.2f} + {coefficients[1]:.2f}*X + {coefficients[2]:.2f}*X^2 + {coefficients[3]:.2f}*X^3 + {coefficients[4]:.2f}*X^4 + {coefficients[5]:.2f}*X^5")
```

```python
# 6. Predicciones
y_pred = model.predict(X_poly)
```

```python
# 7. Evaluar el modelo
mse = mean_squared_error(y, y_pred)
print('Error Cuadr√°tico Medio:', mse)
```

```python
# 8. Graficar los resultados
sorted_indices = X[:, 0].argsort()
X_sorted = X[sorted_indices]
y_sorted = y_pred[sorted_indices]

plt.scatter(X, y, label='Datos reales')      
plt.plot(X_sorted, y_sorted,color = "red", label='Modelo')
plt.title('Datos Sint√©ticos No Lineales y Modelo Polinomial')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
