```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 1. Preparacion de datos

```python
# Cargar el dataset
df = pd.read_csv('../data_mining/Advertising.csv')
df.head()
```

```python
df.describe()
```

```python
# Visualizacion datos a traves de seaborn
sns.pairplot(df, x_vars=['TV', 'radio', 'newspaper'], y_vars='sales', height=5, aspect=0.7)
```

```python
# Cargar los datos
# RL: y =a + bx | ML: y = b0 + b1x
# X (una variable)
#  y: variable objetivo
x = df[['TV']]
y = df['sales']
```

## 2. Separacion en conjunto de entrenamiento y prueba

```python
from sklearn.model_selection import train_test_split
```

```python
# Dividir el dataset en train y test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
# Verificar la forma de los datos
print(f'x_train: {x_train.shape}')
print(f'x_test: {x_test.shape}')
```

## 3. Configuracion del algoritmo

```python
# Regresion lineal
from sklearn.linear_model import LinearRegression
```

```python
# Crear el modelo
model = LinearRegression()
```

## 4. Entrenamiento del modelo

```python
# Ajustar el modelo
model.fit(x_train, y_train)
```

## 5. Prediccion

```python
y_pred = model.predict(x_test)
```

## 6. Evaluacion

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
```

```python
# Verificar el modelo
print(f'Intercepto: {model.intercept_}')
print(f'Pendiente: {model.coef_}')
```

```python
print('Error absoluto medio:', mean_absolute_error(y_test, y_pred))
print('Coeficiente de determinacion R^2:', r2_score(y_test, y_pred))
```

```python
# Calcular el error cuadratico medio
mse = mean_squared_error(y_test, y_pred)
print(f'Error cuadratico medio: {mse}')
```

```python
# Graficar los resultados
sns.scatterplot(x=x_test['TV'], y=y_test, label='Datos reales')
sns.scatterplot(x=x_test['TV'], y=y_pred, label='Datos predichos')
sns.lineplot(x=x_test['TV'], y=y_pred, color='red', label='Linea de predicción')
```
